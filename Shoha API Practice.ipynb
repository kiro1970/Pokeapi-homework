# APIs

#### What are they and how do we use them

<p> API stands for application programming interface - what an API actually does is provide a way for different servers, devices, and applications to send information back and forth in a commonly understood structure. </p>

<p> So using an api, we can send data from an application written in one programming language to an application written in another programming language on a different server, and both applications can understand that data. </p>

<p> Its kind of like the kitchen in a restaurant -> they take the unprepared food items, chop them up, cook them, and lay them out nicely presented on a plate for the customers to consume as they please. </p>

<p> APIs use data formats that can be understood by multiple programming languages -> the most common one we will use is called JSON data (JavaScript Object Notation) </p>

```python
import requests 

class Pokemon:
    def __init__(self, pokedata):
        self.name = pokedata['name']
        self.abilities = pokedata['abilities']
        self.weight = pokedata['weight']
        self.types = pokedata['types']
        
    def printInfo(self):
        print(f'Name: {self.name.capitalize()}')
        for t in self.types:
            n = t['type']['name']
            print("Type: " + n)
        
        for a in self.abilities:
            n = a['ability']['name']
            print("Ability: " + n)
            
        print(f'Weight {self.weight}')
        print()
        
class Pokedex:
    def __init__(self):
        self.dict = {}
        
    def createPokemon(self, starters):
        for name in starters:
            data = requests.get("https://pokeapi.co/api/v2/pokemon/" + name)
            if data.status_code == 200:
                pokedata = data.json()
                p = Pokemon(pokedata)    
                self.dict[p.name] = p
            else:
                print(f'Sorry, failed to add {name}!')
                
    def showAll(self):
        for key in self.dict.keys():
            x = self.dict[key]
            x.printInfo()
    
    def searchType(self, t):
        for value in self.dict.values():
            y = value.types
            for v in y:
                if v['type']['name'] == t:
                    value.printInfo()
                
            
        
team = ["bulbasaur", "charmander", "squirtle", "chikorita", "cyndaquil", "totodile", "treecko", "torchic", "mudkip", "turtwig", "chimchar", "piplup", "snivy", "tepig", "oshawott", "chespin", "fennekin", "froakie", "rowlet", "litten", "popplio", "grookey", "scorbunny", "sobble"] 
dex = Pokedex()
dex.createPokemon(team)
dex.searchType('grass')
dex.searchType('fire')
dex.searchType('water')
```
